## EDA (Exploratory Data Analysis)

Questios that we might have looking from this data:
    - Over the years, the number of sells increase or decresead ?
    - What's the top 10 Games most selled in world ? And in Which region ?
    - Which Publisher selled most games?
    - And which platform released more games?
    - From each Genre, which game was the most popular selled ?

Those are the question that I thought of exploring from this dataset.

```python
plt.figure(figsize = (8,6))
df.groupby('Year')['Global_Sales'].sum().plot(kind = 'line');
plt.title('Game Sells over the years..', fontsize = 12);
```

In this plot, we can observe  that in 1995 was the beginning of the Video Game Era

  - What's the top 10 Games most selled in world ? And in Which region ?

```python
plt.figure(figsize = (8,6))
df.groupby('Name')['Global_Sales'].sum().sort_values(ascending=False)[:10].plot(kind = 'bar', color = 'b', alpha = 0.7)
plt.title('Top 10 Most selled games WORLD', fontsize = 12);
```

```python
plt.figure(figsize = (8,6))
df.groupby('Name')['NA_Sales'].sum().sort_values(ascending=False)[:10].plot(kind = 'bar', color = 'grey', alpha = 0.5)
plt.title('Top 10 Most selled games NA', fontsize = 12);
```

```python
plt.figure(figsize = (8,6))
df.groupby('Name')['EU_Sales'].sum().sort_values(ascending=False)[:10].plot(kind = 'bar', color = 'brown', alpha = 0.5)
plt.title('Top 10 Most selled games EU', fontsize = 12);
```

```python
plt.figure(figsize = (8,6))
df.groupby('Name')['JP_Sales'].sum().sort_values(ascending=False)[:10].plot(kind = 'bar', color = 'blue', alpha = 0.3)
plt.title('Top 10 Most selled games Japan', fontsize = 12);
```

Interesting fact that in Japan Pokemon and Nintendo are very Strong there. In the other hand, NA and Europe we also observe Nintendo Sports but we have GTA. Means that different taste from each continent

- Which Publisher selled most games?

```python
plt.figure(figsize = (8,6))
df.groupby('Publisher')['Global_Sales'].sum().sort_values(ascending = False)[:3].plot(kind = 'barh', color = 'blue', alpha = 0.3)
plt.title('Top sellers publishers', fontsize = 12);
```

Top 3 most sellers publishers are Nintendo, Electronic Arts and Activision. Those three I remember in my childhood

- And which released more games?

```python
plt.figure(figsize = (8,6))
df.groupby('Platform')['Name'].count().sort_values(ascending = False)[:10].plot(kind = 'barh');
plt.title('Platform with more releases', fontsize = 12);
```

- From each Genre, which game was the most selled ?

```python
aux = df.groupby('Genre')[['Global_Sales']].max()

aux.reset_index().merge(df[["Genre", "Global_Sales", "Name"]], 
                        on=["Genre", "Global_Sales"], 
                        how="inner")
```

Sports Game leading. Pro evolution soccer, I prefer Winning eleven haha

Ã“tima ferramenta para identificaÃ§Ã£o de outliers Ã© o boxplot. Com ele podemos visualizar os outliers de forma prÃ¡tica e rÃ¡pida!

Vamos excluir esses outliers...

```python
# para excluir vamos utilizar os Intervalo Interquartilico
Q1 = df['Global_Sales'].quantile(.25)
Q3 = df['Global_Sales'].quantile(.75)
IIQ = Q3 - Q1
limite_inferior = Q1 - 1.5 * IIQ
limite_superior = Q3 + 1.5 * IIQ
```

```python
# filtrando dados somente no intervalo interquartilico
df_woutlier = df[(df['Global_Sales'] >= limite_inferior) & (df['Global_Sales'] <= limite_superior )]
```

```python
df_woutlier.head()
```

```python
# agora vamos plotar o boxplot
sns.boxplot(df_woutlier['Global_Sales'])
```

Bem melhor que o anterior nÃ©, mas ainda temos mt outliers.

```python
sns.histplot(df_woutlier['Global_Sales'])
```

```python
sns.histplot(df_woutlier['Global_Sales'], kde = True)
```

```python
plt.figure(figsize = (12,5))
sns.violinplot(data = df_woutlier, x = "Genre", y = "Global_Sales")
plt.title("Boxplots da vendas por genero de game", fontsize = 12);
plt.show()
```
