```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
ax = dados.sort_values('2008/Ago', ascending = False).plot(x = 'Unidade da Federação', y = '2008/Ago', kind = 'bar', figsize = (10,7))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}') )
plt.title('Estados do Sul e Sudeste lideram em gastos com procedimentos hospitalares', fontsize = 14);
```

```python
ax = dados.sort_values('2021/Mar', ascending = False).plot(x = 'Unidade da Federação', y = '2021/Mar', kind = 'bar', figsize = (10,7))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}') )
plt.title('Valor por unidade por federacao', fontsize = 14);
```
